```python
def value_iteration(env, gamma=0.9, max_iterations=100000):
  n_state = env.observation_space.n
  n_action = env.action_space.n
  n_iter = 0
  value_function = np.zeros(n_state)
  policy = np.zeros(n_state)
  delta = 1
  theta = 10**-5

  while(delta>=theta):
    delta = 0
    for state in range(n_state):
      value_old = value_function[state]
      for action in range(n_action):
        value_new = 0
        for prob, new_state, reward, terminated in env.P[state][action]:
          value_new += prob * (reward + gamma*value_function[new_state])
        value_function[state] = max(value_function[state], value_new)
      delta = max(delta, abs(value_old - value_function[state]))
    n_iter += 1

    for state in range(n_state):
        max_value = float('-inf')
        for action in range(n_action):
            value = 0
            for prob, new_state, reward, terminated in env.P[state][action]:
              value += prob * (reward + gamma*value_function[new_state])
            if (value > max_value):
              max_value = value
              policy[state] = action

  return value_function, n_iter, policy
```

```python
def policy_iteration(env, gamma=0.9, max_iterations=100000):
  n_state = env.observation_space.n
  n_action = env.action_space.n
  n_iter = 0
  value_function = np.zeros(env.observation_space.n)
  policy = [np.random.choice(range(n_action)) for state in range(n_state)]

  while(True):
    delta = 1
    theta = 10**-5

    while(delta>=theta):
      delta = 0
      for state in range(n_state):
        value_old = value_function[state]
        action = policy[state]
        value_new = 0
        for prob, new_state, reward, terminated in env.P[state][action]:
          value_new += prob * (reward + gamma*value_function[new_state])
        value_function[state] = max(value_function[state], value_new)
        delta = max(delta, abs(value_old - value_function[state]))
      n_iter += 1
    
    policy_stable = True
    for state in range(n_state):
      action_old = policy[state]
      action_new = n_action
      max_value = float('-inf')
      for action in range(n_action):
        value = 0
        for prob, new_state, reward, terminated in env.P[state][action]:
          value += prob * (reward + gamma*value_function[new_state])
        if (value > max_value or (value == max_value and action == action_old)):
          max_value, action_new = value, action
      policy[state] = action_new
      if(action_old!=action_new):
        policy_stable = False
 
    if policy_stable:
      break

  return value_function, n_iter
```

```python
def monte_carlo_prediction(env, policy, gamma, iter=10000):
  n_state = env.observation_space.n
  n_action = env.action_space.n
  value_function = np.zeros(n_state)
  count = [0 for state in range(n_state)]

  for _ in range(iter):
    state = env.reset()
    episode = []
    done = False
    n_visit = [0 for i in range(n_state)]
    
    while not(done):
      n_visit[state] += 1
      action = policy[state]
      new_state, reward, bv, done, info = env.step(action)
      episode.append([state, action, reward])
      state = new_state

    G = 0
    for state, action, reward in reversed(episode):
      G = gamma * G + reward
      if n_visit[state] > 1:
        n_visit[state] -= 1
        continue
      count[state] += 1
      value_function[state] = value_function[state] + (G - value_function[state])/count[state]

  return value_function
```

```python
# dimension 1 -> state, dimension 2 -> action
r = np.zeros((16,16))
for i in range(16):
  for j in range(16-i):
    r[i][j] = F[i+j] - C[j] - h[i+j]
    
dbr = pd.DataFrame(r)
print('Reward Matrix\n')
print(dbr)
```

```python
# dimension 1 -> state, dimension 2 ->  action, dimension 3 -> next state
T = np.zeros((16,16,16))
for i in range(16):
  for j in range(16-i):
    for k in range(16):
      if k > i + j:
        T[i][j][k] = 0
      elif k == 0:
        T[i][j][k] = q[i+j]
      else:
        T[i][j][k] = p[i+j-k]
for i in range(16):
  dbt = pd.DataFrame(T[i])
  print('Transition Matrix for state ',i,'\n')
  print(dbt)
```

## Q2) Optimal Policy using Value Iteration

```python
# this cell runs for about 1 min 45 secs

value_function = np.zeros(16)
policy = np.zeros(16)
delta = 1
theta = 10**-2
vf = [] # stores value function at each iteration
max_vf = [] # stores maximum of value function at each iteration
deltas = [] # stores the distance(max norm) between current value function and previous value function in each iteration

while(delta>=theta):
  delta = 0
  for state in range(16):
    value_old = value_function[state]
    for action in range(16-state):
      value_new = 0
      for new_state in range(16):
        value_new += T[state][action][new_state] * (r[state][action] + value_function[new_state])
      value_function[state] = max(value_function[state], value_new)
    delta = max(delta, abs(value_old - value_function[state]))
  vf.append(list(value_function))
  max_vf.append(max(value_function))
  deltas.append(delta)

for state in range(16):
    max_value = float('-inf')
    for action in range(16-state):
        value = 0
        for new_state in range(16):
          value += T[state][action][new_state] * (r[state][action] + value_function[new_state])
        if (value > max_value):
          max_value = value
          policy[state] = action
```
